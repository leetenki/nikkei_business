```python
# csvファイル一覧
import os
import glob

csv_files = sorted(glob.glob("./2020年*.csv"))
csv_files
```

```python
# csvファイル読み込み
import csv
csv_data = []
for csv_file in csv_files:
  with open(csv_file, 'r', encoding="utf-8") as f:
    csv_data += [row for row in csv.reader(f)]

print("%d件の販売履歴" % (len(csv_data)))
csv_data[:10]
```

```python
# 時系列順にソート
csv_data = sorted(csv_data, key=lambda x: x[1])
csv_data[:10]
```

```python
# ユニークなマンション名リストを作成
mansion_names = []
for row in csv_data:
  mansion_name = row[0]
  if not mansion_name in mansion_names:
    mansion_names.append(mansion_name)

print("%d件のマンション名" % (len(mansion_names)))
mansion_names[:10]
```

```python
# マンション名の完全一致するものをまとめて、販売履歴の配列を作成
indexed_mansion_data = {}
for row in csv_data:
  mansion_name = row[0]
  idx = mansion_names.index(mansion_name)
  if not idx in indexed_mansion_data:
    indexed_mansion_data[idx] = []
  indexed_mansion_data[idx].append(row)

indexed_mansion_data
```

```python
# トークン分割関数を定義
from janome.tokenizer import Tokenizer

def tokenize(text):
    t = Tokenizer()
    tokens = t.tokenize(text.replace('・', ' '))
    return [x.surface.strip() for x in tokens]

tokenize("ライオンズTower東京")
```

```python
# マンション名を全てトークン分割
tokenized_mansion_names = []
for mansion_name in mansion_names:
  tokens = tokenize(mansion_name)
  tokens = [token for token in tokens if len(token) > 0] # 空白文字削除
  tokenized_mansion_names.append(tokens)

tokenized_mansion_names[:10]
```

```python
# カナ読みに変換する関数定義
import pykakasi
def to_kana(words):
    kakasi = pykakasi.kakasi()
    kakasi.setMode('H', 'K')
    kakasi.setMode('J', 'K')
    conv = kakasi.getConverter()
    return [conv.do(x) for x in words]

to_kana(['ライオンズ', 'Tower', '東京'])
```

```python
# 全ての漢字トークンを可能な限りかな読みに変換
kana_mansion_tokens = []
for tokens in tokenized_mansion_names:
  kana_mansion_tokens.append(to_kana(tokens))

kana_mansion_tokens[:10]
```

```python
# 英語を日本語カナ読みに変換する関数定義
import requests
import json
import re
import time

def en_to_kana(words):
    base = "https://www.sljfaq.org/cgi/e2k_ja.cgi?o=json&word="
    headers = {'User-Agent': 'Mozilla/5.0' } 
    results = []
    for word in words:
        if not re.match(r'^[a-zA-Z]+$', word):
            results.append(word)
            continue
        url = base + word
        print("converting [%s]..." % (word))
        res = requests.get(url, headers=headers)
        kana = [x["j_pron_spell"] for x in json.loads(res.text)["words"]]
        time.sleep(10) # サーバに負荷をかけないための配慮
        print("%s -> %s" % (word, kana))
        results += kana
    return results

en_to_kana(['ライオンズ', 'Tower', 'トウキョウ'])
```

```python
# 全ての英語トークンをカナ読みに変換
en_kana_mansion_tokens = []
for tokens in kana_mansion_tokens:
  en_kana_mansion_tokens.append(en_to_kana(tokens))
  
en_kana_mansion_tokens[:10]
```

```python
# TF-IDF値に変換
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

docs = [" ".join(x) for x in en_kana_mansion_tokens]
vectorizer = TfidfVectorizer(binary=True,use_idf=False)
vecs = vectorizer.fit_transform(docs)

print(vectorizer.get_feature_names())
print(vecs.toarray())
print(vecs.shape)
```

```python
# コサイン類似度配列作成
similarity = np.round(cosine_similarity(vecs), 3)

print(similarity)
print(similarity.shape)
```

```python
# コサイン類似度0.9以上のものを同一マンションと見なして配列にまとめる
import numpy as np

mansion_names = np.array(mansion_names)
unique_mansion_names = [list(mansion_names[same_mansion_idx]) for same_mansion_idx in (similarity > 0.9)]
unique_mansion_names = list(np.unique(np.array(unique_mansion_names, dtype=object)))

print(len(unique_mansion_names))
unique_mansion_names
```

```python
mansion_name_indices = {}
for id, names in enumerate(unique_mansion_names):
    for name in names:
        mansion_name_indices[name] = id

mansion_name_indices
```

```python
# 名寄せした結果でマンションごとの売買履歴の配列を作成
unique_indexed_mansion_data = {}
for row in csv_data:
  mansion_name = row[0]
  idx = mansion_name_indices[mansion_name]
  if not idx in unique_indexed_mansion_data:
    unique_indexed_mansion_data[idx] = []
  unique_indexed_mansion_data[idx].append(row)

print(len(unique_indexed_mansion_data))
unique_indexed_mansion_data
```

```python
# csv出力
out_csv_data = []
for mansion_id in sorted(unique_indexed_mansion_data.keys()):
    out_csv_data += [[mansion_id] + history for history in unique_indexed_mansion_data[mansion_id]]
    
with open('名寄せ結果.csv', 'w') as f:
    writer = csv.writer(f)
    for row in out_csv_data:
         writer.writerow(row)

out_csv_data[:30]
```
